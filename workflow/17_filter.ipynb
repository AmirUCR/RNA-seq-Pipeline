```python
import pandas as pd

df = pd.read_csv("../data/output/edger_results.csv")

condition = ((df['log2FoldChange'] >= 0.5) | (df['log2FoldChange'] <= -0.5)) & \
            (df['24WT1_H9'] >= 5) & \
            (df['24WT2_A1'] >= 5) & \
            (df['24WT3_H11'] >= 5) & \
            (df['24PKLKO1_A2'] >= 5) & \
            (df['24PKLKO2_H10'] >= 5) & \
            (df['24PKLKO3_H12'] >= 5) & \
            (df['FDR'] < 0.05)

df = df[condition]
df.to_csv("../data/output/edger_filtered_results.csv", index=False)
edger_genes = set(df["name"])
df
```

```python
edger_downreg = df[df["log2FoldChange"] < 0].reset_index(drop=True)
edger_downreg.to_csv("../data/output/edger_filtered_downreg.csv", index=False)
edger_downreg
```

```python
edger_upreg = df[df["log2FoldChange"] > 0].reset_index(drop=True)
edger_upreg.to_csv("../data/output/edger_filtered_upreg.csv", index=False)
edger_upreg
```

```python
df = pd.read_csv("../data/output/deseq2_results.csv")

condition = ((df['log2FoldChange'] >= 0.5) | (df['log2FoldChange'] <= -0.5)) & \
            (df['24WT1_H9'] >= 5) & \
            (df['24WT2_A1'] >= 5) & \
            (df['24WT3_H11'] >= 5) & \
            (df['24PKLKO1_A2'] >= 5) & \
            (df['24PKLKO2_H10'] >= 5) & \
            (df['24PKLKO3_H12'] >= 5) & \
            (df['FDR'] < 0.05)

df = df[condition]
df.to_csv("../data/output/deseq2_filtered_results.csv", index=False)
deseq_genes = set(df["name"])
df
```

```python
inter = deseq_genes.intersection(edger_genes)
print("Num Deseq2 genes", len(deseq_genes))
print("Num EdgeR genes", len(edger_genes))
print(len(inter))
inter
```
