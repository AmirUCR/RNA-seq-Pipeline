```python
import pandas as pd

ALL = "../../data/output/edger_results.csv"
df_all = pd.read_csv(ALL)

gene_col = "Name" if "Name" in df_all.columns else df_all.columns[0]

# All genes you tested (population)
universe = (
    df_all[gene_col]
    .dropna()
    .astype(str)
    .drop_duplicates()
)

# Write ids_pop.txt
universe.to_csv("ids_pop.txt", index=False, header=False)
print(f"Wrote {len(universe)} genes to ids_pop.txt")
```

```python
IN = "../../data/output/edger_filtered_results.csv"
df = pd.read_csv(IN)

fdr_col = "FDR" if "FDR" in df.columns else ("PAdj" if "PAdj" in df.columns else None)

if fdr_col and (df[fdr_col] <= 0.05).any():
    query = df.loc[df[fdr_col] <= 0.05, gene_col].dropna().astype(str).drop_duplicates()
else:
    score_col = "logFC" if "logFC" in df.columns else "log2FoldChange"
    query = (
        df.assign(_abs=df[score_col].abs())
          .sort_values("_abs", ascending=False)[gene_col]
          .dropna()
          .astype(str)
          .drop_duplicates()
        #   .head(300)  # or whatever top-N you use
    )

query.to_csv("ids_stu.txt", index=False, header=False)
print(f"Wrote {len(query)} genes to ids_stu.txt")
```
